```python
print('Sum df Homeruns by yearID against teamID where yearID > 2000 into team_homeruns_per_year_df and fill N/A with 0.0.')
team_homeruns_per_year_df = df[df.yearID > 2000].pivot_table(values='Homeruns', index='yearID', columns='teamID', aggfunc='sum', fill_value=0.0)
team_homeruns_per_year_df.head()
```

```python
print('Plot team_homeruns_per_year_df as an area chart.')
_ = team_homeruns_per_year_df.plot.area()
```

## Improve the revious area chart:
* including only the first 5 teams sorted by the highest number of home runs in any year [2001 - 2015]:
* including only the first 5 teams sorted by the average number of home runs over the years [2001 - 2015]:

### 1) Including only the first 5 teams sorted by the highest number of home runs in any year [2001 - 2015]:

```python
# select all of the teamID which belong to the first 5 teams sorted by the highest number of home runs.
h_df = df[(df.yearID >= 2001) & (df.yearID <= 2015)].groupby(['yearID', 'teamID']).agg({'Homeruns': 'sum'}).reset_index()
h_df['rank'] = h_df['Homeruns'].groupby(h_df.yearID).rank(ascending = 0, method = 'dense').astype(np.int64)
h_df = h_df[h_df['rank'] < 6]
print(h_df)
```

```python
# drop irrelevant data
h_df = h_df['teamID'].unique()
h_df = df.loc[df['teamID'].isin(h_df)]
h_df.head(10)
```

```python
# compute sum of homeruns by yearID and draw the area plot.
print('Sum df Homeruns by yearID against teamID where yearID between 2001-2015 into team_homeruns_per_year_df and fill N/A with 0.0.')
team_homeruns_per_year_df = h_df[(h_df.yearID > 2000) & (h_df.yearID < 2016)].pivot_table(values='Homeruns', index='yearID', columns='teamID', aggfunc='sum', fill_value=0.0)
print(team_homeruns_per_year_df.head())
print(' ')
print('Improved plot team_homeruns_per_year_df as an area chart.')
_ = team_homeruns_per_year_df.plot.area()
```

### 2) including only the first 5 teams sorted by the average number of home runs over the years [2001 - 2015]:

```python
# Select the teams which are the first 5 teams sorted by the average number of home runs.
a_df = df[(df.yearID >= 2001) & (df.yearID <= 2015)].groupby(['yearID', 'teamID']).agg({'Homeruns': 'sum'}).reset_index()
a_df = a_df.groupby(['teamID'])['Homeruns'].mean().reset_index()
print('Mean')
print(a_df)
a_df['rank'] = a_df['Homeruns'].rank(ascending = 0, method = 'dense').astype(np.int64)
a_df = a_df[a_df['rank'] < 6]
a_df = a_df['teamID'].unique()
a_df
```

```python
# Drop irrelevant data
a_df = df.loc[df['teamID'].isin(a_df)]
a_df.head(10)
```

```python
# Computs the sum of Homeruns based on the above selected 5 teams and draw the area plot.
print('Sum df Homeruns by yearID against teamID where yearID between 2001-2015 into team_homeruns_per_year_df and fill N/A with 0.0.')
team_homeruns_per_year_df = a_df[(a_df.yearID > 2000) & (a_df.yearID < 2016)].pivot_table(values='Homeruns', index='yearID', columns='teamID', aggfunc='sum', fill_value=0.0)
print(team_homeruns_per_year_df)
print(' ')
print('Improved plot team_homeruns_per_year_df as an area chart.')
_ = team_homeruns_per_year_df.plot.area()
```
